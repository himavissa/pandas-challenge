```python
import pandas as pd
```

```python
#Read csv file
student_data_file = "students_complete.csv"
student_data_df = pd.read_csv(student_data_file,encoding="utf-8")
student_data_df.head()

schools_data_file = "schools_complete.csv"
schools_data_df = pd.read_csv(schools_data_file,encoding="utf-8")
schools_data_df.head()

```

```python
#Total no of schools
total_schools = schools_data_df["school_name"].count()
total_schools
```

```python
#Total no of students
total_Students = student_data_df["Student ID"].count()
total_Students
```

```python
#Total budget
total_budget = schools_data_df["budget"].sum()
total_budget
```

```python
#Average Math score
avg_mathscore = student_data_df["math_score"].mean()
avg_mathscore
```

```python
#Average Reading Score
avg_reading_score = student_data_df["reading_score"].mean()
avg_reading_score
```

```python
# Passing Math (The percentage of students that passed math.)

student_data_df["#passing_math"] = student_data_df["math_score"] >=70
per_pass_math = ((student_data_df["#passing_math"]).mean()) * 100
per_pass_math



                 
```

```python
#Passing Reading Percentange
student_data_df["#passing_read"] = student_data_df["reading_score"] >=70
per_pass_read = ((student_data_df["#passing_read"]).mean()) * 100
per_pass_read
```

```python
# Overall Passing (The percentage of students that passed math **and** reading.)

overrall_passing_rate = (per_pass_math + per_pass_read)/2
overrall_passing_rate
```

```python
#Create a dataframe to hold the above results
district_results_df = [{"Total Schools": total_schools, 
            "Total Students": total_Students, 
            "Total Budget": total_budget, 
            "Average Math Score":  round(avg_mathscore,2), 
            "Average Reading Score":  round(avg_reading_score,2), 
           "% Passing Math": round(per_pass_math,2),
           "% Passing Reading": round(per_pass_read,2),
            "% Overall Passing Rate": round(overrall_passing_rate,2)}]
district_results_df = pd.DataFrame(district_results_df)
district_results_df
```

```python
#Merge two dataframes into new
complete_school_data_df = pd.merge(student_data_df, schools_data_df, how="left", on=["school_name", "school_name"])
complete_school_data_df.head()
```

```python
#School Summary
#group by school name 
schoolsumm_group_df = complete_school_data_df.groupby(["school_name"]).mean()
#creating new coloms
schoolsumm_group_df["Per Student budget"] = schoolsumm_group_df["budget"]/schoolsumm_group_df["size"]
schoolsumm_group_df["% Passing Math"] = schoolsumm_group_df["#passing_math"] *100
schoolsumm_group_df["% Passing Reading"] = schoolsumm_group_df["#passing_read"] *100
schoolsumm_group_df["% Overall Passing"] = ((schoolsumm_group_df["#passing_math"] + schoolsumm_group_df["#passing_read"])/2)*100

# Merge with school data for type, size, total school budegt
school_data_summdf = pd.merge(schoolsumm_group_df,schools_data_df,how="left",on=["school_name"])
school_data_summdf

#School Summary data frame

Summary_df = pd.DataFrame({"School Name": school_data_summdf["school_name"],
               "School Type": school_data_summdf["type"],
            "Total Students": school_data_summdf["size_x"], 
            "Total School Budget": school_data_summdf["budget_x"],
                          "Per Student Budget":  school_data_summdf["Per Student budget"],
             "Average Math Score"  : round(school_data_summdf["math_score"],2),
            "Average Reading Score":  round(school_data_summdf["reading_score"],2), 
           "% Passing Math": round(school_data_summdf["% Passing Math"],2),
           "% Passing Reading":round(school_data_summdf["% Passing Reading"],2),
            "% Overall Passing Rate":school_data_summdf["% Overall Passing"]})
Summary_df

```

```python
#Display top 5 schools from overall passsing
top_5_df = Summary_df.sort_values(["% Overall Passing Rate"],ascending=False)
top_5_df.head()
```

```python
#Bottom porforming schools
bottom_5_df = Summary_df.sort_values(["% Overall Passing Rate"],ascending=True)
bottom_5_df.head()
```

```python
#Avergae math score of each school(use group by )and thier grades and create series
nineth = complete_school_data_df[complete_school_data_df["grade"] == "9th"].groupby("school_name")["math_score"].mean()
tenth = complete_school_data_df[complete_school_data_df["grade"] == "10th"].groupby("school_name")["math_score"].mean()
eleventh = complete_school_data_df[complete_school_data_df["grade"] == "11th"].groupby("school_name")["math_score"].mean()
twelveth = complete_school_data_df[complete_school_data_df["grade"] == "12th"].groupby("school_name")["math_score"].mean()

#Display in dataframe
Grade_scores_df = pd.DataFrame({"9th" : round(nineth,2),
                               "10th" : round(tenth,2),
                               "11th" : round(eleventh,2),
                               "12th" : round(twelveth,2)})
Grade_scores_df


```
